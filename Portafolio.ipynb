##4. ¿CÓMO SE CREA UNA CLASE ABSTRACTA?

 ### **Clase abstracta**
 1. ### *Es una clase base.*
 2. ### Nos sirve para agrupar atributos y metodos de una serie de clases. 
 3. #### **No podemos definir objectos**. 

```python
%%writefile ClaseAbstracta.cpp

#include<iostream>
using namespace std;

class Figura  // clase abstracta (Podemos diferenciar una clase abstracta por el uso del 0 'significar PURE')
{
  protected:
    int x, y; //atributos generales

  public: 
  
        Figura()
        {
            x = y = 0;
        }

       virtual void Mostrar() = 0; // esta declaracion obliga a las clases hijas a tener el metodo Mostrar()
                                   // ES UNA DECLARACION DE UN METODO COMO PURE
};

class Cuandrado : public Figura 
{
    int l;
 
 public: 
  Cuandrado(int l)
  {
      this->l = l;
  }

  void Mostrar()
  {
      for(int i = 0; i<l; i++)
        {
          for(int j = 0; j<l; j++)
            {
               cout<<"* ";   
            }
           cout<<endl;
        }
  }


 };

int main()
{
    Cuandrado c(5);
 
    c.Mostrar();

    return 0;
}
```

```python
!g++ ClaseAbstracta.cpp -o s && ./s
```

## ¿CÓMO SE ASIGNA LA MEMORIA EN C++?

En C++ la asignación dinámica de memoria se hace con la ayuda de punteros. 


```python
%%writefile MemoriaD.cpp

// new: Nos ayuda a reservar el espacio de memoria requerido. (un numero de bytes)
// delete: Nos sirve para liberar el espacio de memoria reservado anteriormente.   

#include<iostream>
#include<cstdlib> // para el uso de new y delete
#include<ctime>

using namespace std;

int main()
{
  srand(time(0));

  //int array[50]; // 4bytes * 50 = 200 bytes

  int nNum, *numeros; 
 
  cout<<"Cuantos numeros desea generar? ";
  cin>>nNum;
 
  numeros = new int[nNum]; // 4bytes * nNum = 4*nNum bytes
                          // reservamos espacio en memoria

  for(int i = 0; i<nNum; i++)
      numeros[i] = 10 +  rand()%50;
  
  for(int i = 0; i<nNum; i++)
      cout<<numeros[i]<<" ";


  delete[] numeros; // liberamos el espacio de memoria reservado

  return 0;

}
```

#12. ¿Qué es la herencia y cómo lo utilizó en su proyecto?
La herencia, como su mismo nombre lo indica, es la capacidad de una clase base(padre) de heredar sus miembros y funciones a una clase hija. Lo use para ahorrar código, por ejemplo, la clase enemigo heredaba funciones de la clase Personaje.


```python
%%writefile p12.cpp
#include <iostream>
using namespace std;
class Personaje{
    private: 
        string nombre;
        int x;
        int y;
    public:
        Personaje(string,int,int);
        void mostrarPersonaje();
};

class Enemigo : public Personaje{
    private:
        string poder;
        int vidas;
    public:
        Enemigo(string, int,int,string,int);
        void mostrarEnemigo();
};

Personaje::Personaje(string _nombre, int _x,int _y){
    nombre= _nombre;
    x=_x;
    y=_y;
}

Enemigo::Enemigo(string _nombre, int _x, int _y, string _poder, int _vidas) : Personaje(_nombre, _x, _y){
    poder=_poder;
    vidas=_vidas;
}

void Personaje::mostrarPersonaje(){
    cout<<"Nombre: "<<nombre<<endl;
    cout<<"x: "<<x<<endl;
    cout<<"y: "<<y<<endl;
}

void Enemigo::mostrarEnemigo(){
    mostrarPersonaje();
 cout<<"Poder: "<<poder<<endl;
 cout<<"Vidas: "<<vidas<<endl;
}

int main()
{
    Enemigo enemigo1("Berserker",100,100,"500",5);
    enemigo1.mostrarEnemigo();
    return 0;
}
```

4. Buscar un número en un arreglo

```python
%%writefile Buscararreglo.cpp
#include <iostream>
#include<cstdlib>
#include<ctime>
using namespace std;
void DarValores(int* ejemplo, int n) {
    int x = rand() % 50;                                              //4
    for (int i = 0; i < n; i++) {                                     //1+n(1+2+2)+"1"
        ejemplo[i] = x;
    }
}
void BuscarArreglo(int* ejemplo, int Total, int Ubicacion) {
    int aux = 0;                                                      //1
    for (int i = 0; i < Total; i++) {                                 //1+n(1+1+2+2)+"1"
        if (i == Ubicacion)
            aux = ejemplo[i];
    }
    cout << aux;                                                      //1
}
int main() {
    int n = 0;                                                        //1
    int ubicacion;
    cout << "Ingrese el tamaño del arreglo: ";                        //1
    cin >> n;                                                         //1
    int* ejemplo = new int[n];                                        //3
    DarValores(ejemplo, n);
    cout << "Indique la ubicacion del arreglo que quiera ver: ";      //1
    cin >> ubicacion;                                                 //1
    BuscarArreglo(ejemplo, n, ubicacion);
    return 0;
    system("pause");
}

//Detallado = [4 + 1 + n(5) + "1"] + [1 + 1 + n(6) + "1"] + [1 + 1 + 1 + 1 + 3] = 11n + 16
```

5. Calcular el factorial de N

```python
%%writefile FactorialdeN.cpp
#include <iostream>
using namespace std;
void Factorial(int n)
{
	int factorial = 1;                                                  //1
	for (int i = 1; i <= n; i++)                                        //1 + n(1 + 2 + 2)
	{
		factorial = factorial * i;
	}
	cout << "El factorial de " << n << " es " << factorial << endl;     //5
}
int main()
{
	int n = 1;                                                          //1
	cout << "Ingrese un numero: ";                                      //1
	cin >> n;                                                           //1
	Factorial(n);
	return 0;
	system("pause");
}

//Detallado = 2 + 5n + 5 + 3 = 5n + 10
//Big(O) = O(n)
```
